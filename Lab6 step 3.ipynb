```python
def plot_time_series(df, title):
    plt.figure(figsize=(10, 5))
    time_x_axis = np.arange(1, len(df) + 1)
    for col in df.columns:
        if col.lower() != 'current time':
            plt.plot(time_x_axis , df[col], label=col)
    plt.xlabel('Time')
    plt.ylabel('Sensor Value')
    plt.yscale('log')
    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.tight_layout()
    plt.show()


# Plot all sensors in one figure
plot_time_series(outside, 'Sensors Outside Data vs Time')
plot_time_series(roof, 'Roof Data vs Time')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Convert 'Current Time' to datetime
roof['Current Time'] = pd.to_datetime(roof['Current Time'])
outside['Current Time'] = pd.to_datetime(outside['Current Time'])

# Create a numeric time axis for both dataframes
roof_time_axis = np.arange(1, len(roof) + 1)
outside_time_axis = np.arange(1, len(outside) + 1)

sensor_columns = [col for col in roof.columns if col.lower() != 'current time']
for sensor in sensor_columns:
    plt.figure(figsize=(10, 5))
    
    # Plot for Roof
    plt.plot(roof_time_axis, roof[sensor], label=f'Roof {sensor}')
    
    # Plot for Outside
    plt.plot(outside_time_axis, outside[sensor], label=f'Outside {sensor}')
    
    plt.xlabel('Time')
    plt.ylabel(sensor)
    plt.title(f'{sensor} vs Time (Roof vs Outside)')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```

```python
sensor_columns = [col for col in roof.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    roof_data = roof[sensor].dropna()
    outside_data = outside[sensor].dropna()

    plt.figure(figsize=(10, 5))
    plt.hist(roof_data, bins=50,  label=f'roof {sensor}')
    plt.hist(outside_data, bins=70,  label=f'Outside {sensor}')
    plt.xlabel(sensor)
    plt.ylabel('Frequency')
    plt.title(f'{sensor} Distribution: Roof vs Outside')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

How do the data sets compare?

For the higher particles their data overlaps meaning that they similar in distribution. The other data we can see more spread meaning that there are differences in variability. 

```python
sensor_columns = [col for col in roof.columns if col.lower() != 'current time']

for sensor in sensor_columns:
    roof_data = roof[sensor].dropna()
    outside_data = outside[sensor].dropna()

    mu_in = np.mean(roof_data)
    mu_out = np.mean(outside_data)

    sigma_in = np.std(roof_data, ddof=1)
    sigma_out = np.std(outside_data, ddof=1)

    n_rf = len(roof_data)
    n_out = len(outside_data)

    SE_rf = sigma_in / np.sqrt(n_rf)
    SE_out = sigma_out / np.sqrt(n_out)

    SE_delta = np.sqrt(SE_rf**2 + SE_out**2)

    delta = abs(mu_in - mu_out)

    if delta > 3 * SE_delta:
        print(f"For sensor {sensor}, the means are separated by more than 3σ; statistically significant difference.")
    else:
        print(f"For sensor {sensor}, the means are not separated by more than 3σ; no significant difference.")
```
